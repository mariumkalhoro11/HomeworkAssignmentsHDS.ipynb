## **Question 2: Determine the data type of 1+2 and 1+2. Explain why they are different**

```python
# Question 2

print(type(1+2))
print(type(1+2.))
```

1+2 returns an integer (int), while 1+2. returns a floating point number (float).  

This is because 2. includes a decimal point, so Python treats it as a float. When an integer and a float are used together, Python converts the result to a float.

## **Question 4: Create a loop that prints i² starting from i = 1 and stops when i² exceeds 1000. Print the final value of i and explain why i = 32**

```python
# Question 4

i = 1

while i**2 < 1000:
    print(i**2)
    i += 1

print(i)
```

The final value of i is 32.

This is because:
31² = 961, which is still less than 1000  
32² = 1024, which is greater than 1000  

The loop stops when i² exceeds 1000, so the final value of i is 32.
