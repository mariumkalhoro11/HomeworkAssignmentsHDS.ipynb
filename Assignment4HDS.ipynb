```python
## **Question 1: What is the difference between a spreadsheet and a database?**

Spreadsheet: A file like Excel where data is stored in rows and columns. It is best for small datasets, calculations, and quick analysis.

Database: A structured system that stores large amounts of data in multiple related tables. It is designed for efficiency, accuracy, and multiple users.

**Key difference:** Spreadsheets are simple and flexible, while databases are more powerful and designed for large-scale data management.
```

```python
## **Question 2: What are the advantages and disadvantages of using a spreadsheet for health data management?**

**Advantages:**
- Easy to use  
- Good for small datasets  
- Built-in formulas and charts  
- Quick data entry and visualization  

**Disadvantages:**
- Prone to human error  
- Limited security  
- Not ideal for large datasets  
- Difficult for multiple users  
- Poor data integrity (duplicates, inconsistencies)  
```

```python
## **Question 3: What are the advantages and disadvantages of using a database for health data management?**

**Advantages:**
- Handles large datasets efficiently  
- Strong security (important for patient confidentiality)  
- Supports multiple users simultaneously  
- Maintains data accuracy  
- Can link related data (patients, visits, treatments)  

**Disadvantages:**
- Requires technical knowledge  
- More expensive to set up  
- Less flexible for quick analysis  
- Needs structured design  
```

```python
## **Question 4: What are some examples of tasks that can be performed using a spreadsheet?**

- Creating charts and graphs (e.g., infection rates)  
- Budget tracking for a clinic  
- Calculating averages (e.g., patient age)  
- Simple data entry (e.g., vaccination records)  
```

```python
## **Question 5: What are some examples of tasks that can be performed using a database?**

- Managing electronic health records (EHRs)  
- Tracking patient histories across visits  
- Running complex queries (e.g., patients with multiple conditions)  
- Managing hospital systems (appointments, billing)  
```

```python
## **Question 6: What are some examples of tasks that can be performed using both a spreadsheet and a database?**

- Data analysis  
- Reporting (e.g., monthly patient counts)  
- Data storage  
- Sorting and filtering data  
```

```python
## **Question 7: What are some examples of tasks that can be performed using neither a spreadsheet nor a database?**

- Performing surgery  
- Diagnosing patients directly  
- Conducting physical exams  
- Providing patient counseling  

These tasks require human interaction and cannot be performed by data tools.
```

```python
## **Question 8: What are some examples of tasks that can be performed using a spreadsheet but not a database?**

- Quick one-time calculations  
- Creating simple charts instantly  
- Ad-hoc data exploration without setup  
- Personal tracking (e.g., daily calorie log)  
```

```python
## **Question 9: What are some examples of tasks that can be performed using a database but not a spreadsheet?**

- Managing millions of patient records  
- Enforcing strict data security rules  
- Handling multiple users editing data at the same time  
- Performing complex relational queries across multiple tables  
```
