# **Task 1: Henrietta Lacks and HeLa Cells

Henrietta Lacks was a Black woman treated for cervical cancer at Johns Hopkins Hospital in 1951. During her treatment, doctors took samples of her cancer cells without telling her or asking for her consent. These cells later became known as HeLa cells, the first human cells that could survive and reproduce indefinitely in a laboratory.

HeLa cells became extremely important for medical research and were used to develop vaccines, cancer treatments, and many scientific discoveries. However, Henrietta Lacks and her family were never informed about the use of her cells, and they did not receive any compensation or recognition for decades. This raised serious ethical concerns about consent, exploitation, and the rights of patients, especially marginalized communities.

The main ethical issues in this case include the lack of informed consent, violation of patient autonomy, and racial and social injustice. Henrietta Lacks’ cells were taken at a time when Black patients were often treated unfairly in healthcare, and her family only learned about the widespread use of her cells years later. This case shows how researchers can benefit from biological data while the individuals providing it are ignored or harmed.

To prevent this from happening, researchers should have obtained informed consent from Henrietta Lacks before collecting and using her cells. Clear policies about data and tissue ownership should be enforced, and patients should be informed about how their biological materials will be used. Additionally, ethical review boards and strict regulations can help ensure that researchers respect participants’ rights and prevent exploitation in future studies.

# **Task 2: Premera Blue Cross Data Breach**

Another major example of a health data security breach happened in 2015 and involved Premera Blue Cross, a large health insurance company in the United States. In March 2015, Premera announced that hackers had broken into its computer systems and accessed the personal information of about 11 million people. The stolen data included names, addresses, dates of birth, Social Security numbers, bank account information, and medical claim details. The cyberattack lasted for several months before it was discovered, which shows how difficult it can be for organizations to detect advanced hackers. The attackers used malware to access internal systems and copy sensitive information without being noticed. This breach exposed millions of individuals to risks such as identity theft, financial fraud, and misuse of their medical information.

The consequences of the Premera breach were serious. Premera agreed to a $74 million settlement in 2020 to compensate affected individuals and to improve its cybersecurity practices. The company also faced criticism for not having strong enough security measures in place to protect sensitive health data. This incident increased public concern about how health organizations store and protect personal data.

To prevent this breach, Premera could have implemented stronger cybersecurity controls, such as encryption, network monitoring, and regular security audits. Multi-factor authentication and employee training to detect phishing attacks could also have reduced the risk. Overall, the case shows the importance of proactive cybersecurity measures and constant monitoring to protect health data from unauthorized access.

Source: Washington State Attorney General report on the Premera breach — Attorney General investigation into Premera data breach

# **Task 3: Bias in Health Data – Race-Based Kidney Function Algorithms (eGFR)**

Another important example of bias in health data involves how kidney function has historically been measured using race-based formulas. For many years, doctors used an equation called eGFR (estimated glomerular filtration rate) to assess how well a patient’s kidneys were working. This equation included a correction factor for Black patients, assuming that Black individuals naturally have higher muscle mass than other groups. This assumption caused bias because it often overestimated kidney function in Black patients. As a result, Black patients were less likely to be diagnosed with chronic kidney disease early, referred to specialists, or placed on transplant waiting lists. This contributed to racial disparities in kidney care and outcomes.

The consequences of this bias were serious. Many Black patients experienced delayed treatment, worsening disease, and unequal access to kidney transplants. The use of race as a biological factor also reinforced stereotypes and ignored social and environmental factors that influence health. To prevent this bias, researchers and clinicians should use race-neutral algorithms and validate medical models using diverse populations. Instead of using race as a biological variable, doctors should focus on genetic, social, and environmental factors. Increased representation in clinical research and continuous evaluation of medical algorithms can help reduce bias in health data and improve health equity.

Source: New England Journal of Medicine (NEJM) — https://www.nejm.org/doi/full/10.1056/NEJMoa2102953

# **Task 4: Anonymization of Health Data – COVID-19 Data Sharing by the NHS (UK)**

A real example of anonymization of health data happened during the COVID-19 pandemic when the UK National Health Service (NHS) shared large datasets with researchers to study the virus. Because the data included sensitive health information, the NHS used anonymization techniques to protect patient privacy. One major technique used was pseudonymization, where personal identifiers such as names, NHS numbers, and addresses were replaced with unique codes. This allowed researchers to track patient outcomes without knowing their identities. The NHS also used data aggregation and generalization, such as grouping patients by age ranges and regions instead of sharing exact ages or addresses. In some cases, secure research environments were used so researchers could analyze data without downloading it.

These anonymization methods helped protect individual privacy while still allowing scientists to study COVID-19 trends, treatment outcomes, and risk factors. However, there is still a risk of re-identification if anonymized data is combined with other datasets, so strong security policies and limited access were necessary. Overall, this example shows how anonymization techniques like pseudonymization and aggregation can balance privacy protection with the need for public health research.

1. What is the primary goal of data anonymization?

	- The pulse oximeter controversy refers to evidence that these devices are less accurate for individuals with darker skin tones, which can lead to undetected low oxygen levels and contribute to health disparities.

5. What is the role of an Institutional Review Board (IRB) in research?
	- An Institutional Review Board (IRB) reviews research involving human subjects to ensure ethical standards are met, risks are minimized, informed consent is obtained, and participants’ rights and welfare are protected.

6. What is the difference between spreadsheet software and database software for health data management?
	- Spreadsheet software is designed for small-scale data entry and analysis, while database software is built to handle large, complex datasets, support multiple users, enforce data integrity, and provide stronger security and query capabilities.

7. What is a conflict of interest in public health research?
	- A conflict of interest occurs when a researcher’s personal, financial, or professional interests could influence, or appear to influence, the design, conduct, or interpretation of research.

8. What is informed consent in the context of health data collection?
	- Informed consent is the process of providing individuals with clear information about how their health data will be collected, used, stored, and shared, and obtaining their voluntary agreement to participate.

9. What is data masking in the context of de-identification of health data?
	- Data masking is a technique that replaces real personal identifiers (such as names or social security numbers) with altered or fictitious values to prevent identification of individuals.
