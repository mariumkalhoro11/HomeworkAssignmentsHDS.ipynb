```python
import numpy as np
import scipy.stats as stats

# Population
np.random.seed(28)
dBP = np.random.normal(85, 10, 100000)

# Set seed for assignment
np.random.seed(12)
```

## **Question 1: Select a random sample of 40 individuals from the population**

```python
sample = np.random.choice(dBP, 40)
sample
```

## **Question 2: Calculate the sample mean**

```python
sample_mean = sample.mean()
sample_mean
```

## **Question 3: Calculate the sample standard deviation**

```python
sample_std = sample.std(ddof=1)
sample_std
```

## **Question 4: Calculate the standard error**

```python
standard_error = sample_std / np.sqrt(40)
standard_error
```

## **Question 5: Calculate the confidence coefficient for a 95% confidence level**

```python
c = stats.t.ppf(0.975, df=39)
c
```

## **Question 6: Calculate the lower bound of the confidence interval**

```python
lower_bound = sample_mean - c * standard_error
lower_bound
```

## **Question 7: Calculate the upper bound of the confidence interval**

```python
upper_bound = sample_mean + c * standard_error
upper_bound
```

## **Question 8: Create 1000 bootstrapped sample means and sort them**

```python
bootstrapped_means = []

for _ in range(1000):
    boot_sample = np.random.choice(sample, size=40, replace=True)
    bootstrapped_means.append(boot_sample.mean())

bootstrapped_means = np.array(bootstrapped_means)
bootstrapped_means.sort()

bootstrapped_means
```

## **Question 9: Show the 2.5th percentile of the bootstrapped means**

```python
np.percentile(bootstrapped_means, 2.5)
```

## **Question 10: Show the 97.5th percentile of the bootstrapped means**

```python
np.percentile(bootstrapped_means, 97.5)
```
